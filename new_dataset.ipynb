```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold
import numpy as np

def build_pipeline(X):
    numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = X.select_dtypes(include=['object']).columns.tolist()
    
    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer([
        ('num', numeric_pipeline, numeric_features),
        ('cat', categorical_pipeline, categorical_features)
    ])

    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', LogisticRegression(max_iter=1000))
    ])

    return pipeline

# Tria quin conjunt vols avaluar
X_selected = X_static_savry  # o X_savry o X_static_savry

# Per guardar els resultats de cada repetició
all_scores = []

# Fer 50 repeticions
for seed in range(50):
    cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
    pipeline = build_pipeline(X_selected)
    scores = cross_val_score(pipeline, X_selected, y, cv=cv, scoring='roc_auc')
    all_scores.append(scores)

# Convertir a numpy array per facilitar càlculs
all_scores = np.array(all_scores)

# Mitjana i desviació
mean_score = all_scores.mean()
std_score = all_scores.std()

print(f"AUC-ROC mitjà en 50 repeticions de 10-Fold: {mean_score:.3f} ± {std_score:.3f}")
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from enns2 import enns_classification


class ENNSFeatureSelector(BaseEstimator, TransformerMixin):
    def __init__(self, n_bootstraps=20, importance_threshold=0.65, min_selection_ratio=0.7):
        self.n_bootstraps = n_bootstraps
        self.importance_threshold = importance_threshold
        self.min_selection_ratio = min_selection_ratio

    def fit(self, X, y):
        # Convertir a dense array si cal
        X_array = X.toarray() if hasattr(X, "toarray") else (X.values if hasattr(X, "values") else X)
        y_array = y.values if hasattr(y, "values") else y

        # Aplicar ENNS
        selected_idx = enns_classification(X_array, y_array,
                                           n_bootstraps=self.n_bootstraps,
                                           importance_threshold=self.importance_threshold,
                                           min_selection_ratio=self.min_selection_ratio)

        # Guardar les característiques seleccionades
        self.selected_idx_ = selected_idx
        self.n_selected_features_ = len(selected_idx)

        # 🔄 Print automàtic
        print(f"🔍 ENNS ha seleccionat {self.n_selected_features_} característiques.")

        return self

    def transform(self, X):
        if hasattr(X, "toarray"):  # Si és matriu esparsa
            X = X.toarray()
        elif hasattr(X, "iloc"):  # Si és DataFrame
            X = X.values
        return X[:, self.selected_idx_]

    @property
    def n_features_(self):
        return getattr(self, 'n_selected_features_', None)



def build_pipeline(X):
    numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = X.select_dtypes(include=['object']).columns.tolist()
    
    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer([
        ('num', numeric_pipeline, numeric_features),
        ('cat', categorical_pipeline, categorical_features)
    ])

    pipeline = Pipeline([
        ('preprocessor', preprocessor),                 # ✅ Primer preprocessem (convertim a numèric)
        ('feature_selection', ENNSFeatureSelector()),   # ✅ Després seleccionem les millors característiques
        ('classifier', LogisticRegression(max_iter=1000))
    ])

    return pipeline

 
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import roc_auc_score
from sklearn.utils import resample
from joblib import Parallel, delayed
import numpy as np
from enns2 import enns_classification

# Codificar y
y_numeric = y.replace({'Sí': 1, 'No': 0}).astype(int)
scoring_metric = 'roc_auc'

X_selected = X_static_savry  # o X_savry o X_static_savry

def run_cv(seed):
    print(f"🔄 Executant repetició amb seed={seed}...")
    cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
    pipeline = build_pipeline(X_selected)
    scores = cross_val_score(pipeline, X_selected, y_numeric, cv=cv,
                             scoring=scoring_metric, n_jobs=1)
    return scores

# 50 repeticions en paral·lel
all_scores = Parallel(n_jobs=-1, verbose=10)(
    delayed(run_cv)(seed) for seed in range(12)
)

all_scores = np.array(all_scores)
mean_score = all_scores.mean()
std_score = all_scores.std()

print(f"\n🔄 {scoring_metric} global en 50 repeticions de 10 folds: {mean_score:.3f} ± {std_score:.3f}")
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

def build_pipeline_with_rfe(X, n_features_to_select=30):
    numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = X.select_dtypes(include=['object']).columns.tolist()
    
    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer([
        ('num', numeric_pipeline, numeric_features),
        ('cat', categorical_pipeline, categorical_features)
    ])

    # Model base per fer la selecció
    base_estimator = LogisticRegression(max_iter=1000)

    # RFE com a pas intermedi
    rfe = RFE(estimator=base_estimator, n_features_to_select=n_features_to_select)

    pipeline = Pipeline([
        ('preprocessor', preprocessor),  # ✅ Primer preprocessat
        ('feature_selection', rfe),      # ✅ Després RFE per seleccionar les millors característiques
        ('classifier', LogisticRegression(max_iter=1000))  # ✅ Classificador final
    ])

    return pipeline

```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

def build_pipeline_with_rfe_rf_classifier(X, n_features_to_select=30):
    numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = X.select_dtypes(include=['object']).columns.tolist()

    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer([
        ('num', numeric_pipeline, numeric_features),
        ('cat', categorical_pipeline, categorical_features)
    ])

    # Model base per a la selecció de característiques
    base_estimator = LogisticRegression(max_iter=1000)

    # RFE per seleccionar característiques
    rfe = RFE(estimator=base_estimator, n_features_to_select=n_features_to_select)

    # Classificador final: Random Forest
    classifier = RandomForestClassifier(n_estimators=250, random_state=42)

    pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('feature_selection', rfe),
        ('classifier', classifier)
    ])

    return pipeline
```

```python
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier

def build_pipeline_with_rfe_and_smote(X, model=None, n_features_to_select=30):
    if model is None:
        model = LogisticRegression(max_iter=1000)

    numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_features = X.select_dtypes(include=['object']).columns.tolist()

    numeric_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='mean')),
        ('scaler', StandardScaler())
    ])

    categorical_pipeline = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer([
        ('num', numeric_pipeline, numeric_features),
        ('cat', categorical_pipeline, categorical_features)
    ])

    rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)

    pipeline = ImbPipeline([
        ('preprocessor', preprocessor),
        ('feature_selection', rfe),
        ('smote', SMOTE(random_state=42)),
        ('classifier', model)
    ])

    return pipeline
```
