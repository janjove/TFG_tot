```python
## vull fer un histograma del fitcer variables.csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from funcions_net import *


# Llegir el fitxer
df = pd.read_csv('variables.csv', sep=';')

## elimnem les files on utilitat no sigui SI
"df = df[df['Utilitat'] == 'SI']
print(df.size)"
## creem un histograma de la moda de la columna Que és
## fem un for que recorri la coluna que es
```

```python
incognita = 0
PT = 0
item = 0
preg = 0
mitjanes = 0
prob = 0

for i in range(0, len(df['Que és'])):
    if df['Que és'][i] == 'Mitjanes' or df['Que és'][i] == 'Mitjanes(ARG)':
        mitjanes += 1
    elif df['Que és'][i] == 'PREG':
        preg += 1
    elif df['Que és'][i] == 'PT':
        PT += 1
    elif df['Que és'][i] == 'Item':
        item += 1
    elif df['Que és'][i] == 'Prob':
        prob += 1   
    else:
        incognita += 1

## fem un histograma
plt.bar(['Mitjanes', 'PREG', 'PT', 'Item', 'Prob', 'Incognita'], [mitjanes, preg, PT, item, prob, incognita])
plt.ylabel('Nombre de variables')
plt.title('Variables')
plt.show()
```
